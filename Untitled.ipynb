```python
import networkx as nx
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Create a simple knowledge graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])

# Create features for each node
for node in G.nodes():
    G.nodes[node]['feature'] = node * 2

# Split the graph into training and test sets
train_nodes, test_nodes = train_test_split(list(G.nodes()), test_size=0.2)

# Create the training and test sets
train_X = [G.nodes[node]['feature'] for node in train_nodes]
train_y = [node > 3 for node in train_nodes]
test_X = [G.nodes[node]['feature'] for node in test_nodes]
test_y = [node > 3 for node in test_nodes]

# Train a logistic regression model on the training set
clf = LogisticRegression()
train_X = np.array(train_X).reshape(-1, 1)
test_X = np.array(test_X).reshape(-1, 1)
clf.fit(train_X, train_y)

# Test the model on the test set
accuracy = clf.score(test_X, test_y)
print('Accuracy:', accuracy)
```
